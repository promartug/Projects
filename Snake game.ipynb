```python
import pygame
import random

# Initialize pygame
pygame.init()

# Define color constants
BLACK = (0, 0, 0)
GREEN = (0, 128, 0)  # Snake color
RED = (255, 0, 0)    # Food color
GRAY = (25, 25, 25)  # Grid lines color

# Define game dimensions and cell size
WIDTH, HEIGHT = 640, 480
CELL_SIZE = 20

# Define movement directions
LEFT = (-1, 0)
RIGHT = (1, 0)
UP = (0, -1)
DOWN = (0, 1)

# Initialize screen and clock
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Snake Game")
clock = pygame.time.Clock()

# Function to draw the snake segments
def draw_snake(snake):
    for segment in snake:
        pygame.draw.rect(screen, GREEN, pygame.Rect(segment[0] * CELL_SIZE, segment[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE))

# Function to draw the food
def draw_food(food):
    pygame.draw.rect(screen, RED, pygame.Rect(food[0] * CELL_SIZE, food[1] * CELL_SIZE, CELL_SIZE, CELL_SIZE))

# Function to draw the grid lines
def draw_grid():
    for x in range(0, WIDTH, CELL_SIZE):
        pygame.draw.line(screen, GRAY, (x, 0), (x, HEIGHT))
    for y in range(0, HEIGHT, CELL_SIZE):
        pygame.draw.line(screen, GRAY, (0, y), (WIDTH, y))

# Main game loop
def main():
    snake = [(5, 5), (4, 5), (3, 5)]  # Initial snake position
    direction = RIGHT  # Initial direction
    food = (random.randint(0, (WIDTH // CELL_SIZE) - 1), random.randint(0, (HEIGHT // CELL_SIZE) - 1))  # Initial food position
    running = True

    while running:
        screen.fill(BLACK)  # Fill the screen with black background

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            if event.type == pygame.KEYDOWN:
                # Check arrow key inputs to change direction
                if event.key == pygame.K_LEFT and direction != RIGHT:
                    direction = LEFT
                elif event.key == pygame.K_RIGHT and direction != LEFT:
                    direction = RIGHT
                elif event.key == pygame.K_UP and direction != DOWN:
                    direction = UP
                elif event.key == pygame.K_DOWN and direction != UP:
                    direction = DOWN

        head = snake[0]
        new_head = ((head[0] + direction[0]) % (WIDTH // CELL_SIZE), (head[1] + direction[1]) % (HEIGHT // CELL_SIZE))

        # Check for collision with the snake's body
        if new_head in snake:
            running = False

        snake = [new_head] + snake

        # If snake eats food, generate new food. Otherwise, remove the tail.
        if new_head == food:
            food = (random.randint(0, (WIDTH // CELL_SIZE) - 1), random.randint(0, (HEIGHT // CELL_SIZE) - 1))
        else:
            snake.pop()

        draw_grid()  # Draw the grid before drawing snake and food
        draw_snake(snake)
        draw_food(food)

        pygame.display.flip()  # Update the display
        clock.tick(15)  # Control the game's frame rate

    pygame.quit()  # Quit pygame when the game ends

if __name__ == "__main__":
    main()
```
